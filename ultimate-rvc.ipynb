# Download & Extract Models

```python
MODELS_DIR  = f"/kaggle/working/RVC/models/rvc/voice_models"
GDRIVE_ID   = "18nbOpIVOH9YdHN-ARx_ODNC1KB9e-hD5"


try:
    gdown.download_folder(
        id=GDRIVE_ID,
        output=MODELS_DIR,
        quiet=False,
        remaining_ok=True,
    )
    print("  ✅ Models Downloaded.")
except Exception as e:
    print(e)

print("  Unzipping...")
for zf in glob.glob(f"{MODELS_DIR}/**/*.zip", recursive=True) + glob.glob(f"{MODELS_DIR}/*.zip"):
    dest = os.path.join(os.path.dirname(zf), Path(zf).stem)
    if not os.path.exists(dest):
        try:
            os.makedirs(dest, exist_ok=True)
            with zipfile.ZipFile(zf, "r") as z: z.extractall(dest)
            print(f"  ✅ {Path(zf).stem}")
            os.remove(zf)
        except Exception as e: print(f"  ❌ {zf}: {e}")
```

# Frp Configuration

```python
%cd /kaggle/working

import random
import string

chars = string.ascii_letters + string.digits

FRP_NAME = ''.join(random.choice(chars) for _ in range(16))
FRP_SERVER_ADDR = 'XXX_GLOBAL_FRP_SERVER_ADDR_XXX'
FRP_AUTH_TOKEN = 'XXX_GLOBAL_FRP_AUTH_TOKEN_XXX'
FRP_PORT = "XXX_GLOBAL_FRP_PORT_XXX"

#Download Frp
!wget https://github.com/fatedier/frp/releases/download/v0.68.1/frp_0.68.1_linux_amd64.tar.gz
!tar -xzf frp_0.68.1_linux_amd64.tar.gz

#Frp client config
frpc_config = f"""
serverAddr = "{ FRP_SERVER_ADDR }"
serverPort = 7000
auth.token = "{ FRP_AUTH_TOKEN }"

[[proxies]]
name = "{ FRP_NAME }"
type = "tcp"
localPort = 7860
remotePort = { FRP_PORT }
"""
with open('frp_0.68.1_linux_amd64/frpc.toml', 'w') as f:
    f.write(frpc_config)

clear_output()
```
